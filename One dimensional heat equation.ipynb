# One Dimensional Heat Equation

The one-dimensional heat equation describes how temperature in a one-dimensional object changes over time. It is a partial differential equation that is commonly used to model heat conduction in a rod or wire. The equation is given by:





```python
import numpy as np
import matplotlib.pyplot as plt

def true_solution(x, t):
    return np.sin(np.pi * x) * np.exp(-t)

num_space_points = 100
num_time_points = 100
x = np.linspace(0, 1, num_space_points)
t = np.linspace(0, 1, num_time_points)
x, t = np.meshgrid(x, t)
u_true = true_solution(x, t)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, t, u_true, cmap='viridis')
ax.set_xlabel('Space (x)')
ax.set_ylabel('Time (t)')
ax.set_zlabel('Temperature (u)')
ax.set_title('True Solution')
plt.show()
```

The true solution definition above defines the true solution to the one-dimensional heat equation. It takes spatial coordinate x and time coordinate t as input and returns the temperature u at that point.

As an example I generate synthetic data for input and output pairs to help plot and visualize the problem. Here, I specify the number of spatial (num_space_points) and temporal (num_time_points) points. As we know, the np.linspace function is just used to create linearly spaced vectors for spatial (x) and temporal (t) coordinates. np.meshgrid is then used to create a grid of spatial and temporal coordinates. The true solution is evaluated at each point in the grid.

To try different scenarios with the simple heat equation code, you can experiment with: changing the true solution or adjusting the grid resolution. You can also try adding noise to the synthetic data, modify plotting parameters, extend simulation time, etc.

Some examples:

```python
import numpy as np
import matplotlib.pyplot as plt

def modified_true_solution(x, t):
    return np.sin(2 * np.pi * x) * np.exp(-0.5 * t)

# synthetic data with the modified true solution
num_space_points = 100
num_time_points = 100
x = np.linspace(0, 1, num_space_points)
t = np.linspace(0, 1, num_time_points)
x, t = np.meshgrid(x, t)
u_true_modified = modified_true_solution(x, t)


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, t, u_true_modified, cmap='viridis')
ax.set_xlabel('Space (x)')
ax.set_ylabel('Time (t)')
ax.set_zlabel('Temperature (u)')
ax.set_title('Modified True Solution')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


def true_solution(x, t):
    return np.sin(np.pi * x) * np.exp(-t)

# Add noise to the data
def add_noise(data, noise_level=0.05):
    noise = noise_level * np.random.normal(size=data.shape)
    return data + noise

# synthetic data with added noise
num_space_points = 100
num_time_points = 100
x = np.linspace(0, 1, num_space_points)
t = np.linspace(0, 1, num_time_points)
x, t = np.meshgrid(x, t)
u_true = true_solution(x, t)
u_noisy = add_noise(u_true)


fig = plt.figure(figsize=(10, 5))


ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(x, t, u_true, cmap='viridis')
ax1.set_xlabel('Space (x)')
ax1.set_ylabel('Time (t)')
ax1.set_zlabel('Temperature (u)')
ax1.set_title('True Solution')

ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(x, t, u_noisy, cmap='viridis')
ax2.set_xlabel('Space (x)')
ax2.set_ylabel('Time (t)')
ax2.set_zlabel('Temperature (u)')
ax2.set_title('Noisy Data')

plt.tight_layout()
plt.show()
```
